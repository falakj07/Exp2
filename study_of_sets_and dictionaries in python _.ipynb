## Python Dictionaries
Dictionaries are used to store data values in key:value pairs.
 Its a collection of ordered , changeble and do not allow duplicates .

```python
mydict = {
    "career" : "developer",
    "language" : "python ",
    "year" : "2026"}
mydict
```

```python
## Dictionary takes latest value given for key. It will not allow duplicates
mydict = {
    "brand" : "ford",
    "model" : "mustang",
    "year" : "1964 }
    mydict
```

```python
## Changing/Adding  values in dictionaries
car  = {
    "brand" : "hyundai",
    "model" : "i20",
    "year" : "1964"}


car["color"] = "white"
car

```

```python
student = {
    "prn": "25070123047",
    "name" : "falak jain",
    "branch":"E&TC",
}
print("Name:",student["name"])
```

```python
## Adding and Updating Elements
student["year"] = "first"
student["name"] = "Gitika"
print(student)
```

```python
# Removing Elements
student.pop("branch")
print("After removal:",student)
```
